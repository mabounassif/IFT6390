```python
import math
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train_set.csv')
test = pd.read_csv('../data/test.csv')
```

```python
train.shape, test.shape
```

```python
dtypes = train.dtypes
category_columns = dtypes[lambda x: x == 'object']
```

```python
train['source'] = 'train'
test['source'] = 'test'
data = pd.concat([train, test], ignore_index=True)
data.shape
```

```python
data.apply(lambda x: sum(x.isnull()))[lambda x: x > 0]
```

```python
for category_column in category_columns.keys():
    print('\nFrequency count for variable {0}'.format(category_column))
    print(data[category_column].value_counts())
    
    not_na_elements = len(data[category_column].dropna().values)
    na_elements = len(data[category_column]) - not_na_elements
    print('Total: {0} of which {1} are not a number'.format(not_na_elements, na_elements))
```

```python
data = pd.get_dummies(data)
data = data.fillna(data[:data.shape[0]].mean())
data.head()
```

```python
prices = data['SalePrice']

data.drop('SalePrice',axis=1,inplace=True)
data = pd.concat([data, prices], axis=1)
```

```python
train = data.loc[data['source_train'] == 1]
test = data.loc[data['source_test'] == 1]
```

```python
train.drop('source_test',axis=1,inplace=True)
train.drop('source_train',axis=1,inplace=True)

test.drop('source_test',axis=1,inplace=True)
test.drop('source_train',axis=1,inplace=True)
test.drop('SalePrice',axis=1,inplace=True)
```

```python
train.to_csv('train_modified.csv', index=False)
test.to_csv('test_modified.csv', index=False)
```
